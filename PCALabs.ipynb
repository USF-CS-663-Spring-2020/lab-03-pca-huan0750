# WDBC Data

```python
import pandas as pd
```

```python
df=pd.read_csv('./wdbc.data',header=None)
```

```python
df
```

```python
df.describe(include='all')
```

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']

df.columns = column_labels
df
```

```python
df.diagnosis.unique()
df.diagnosis.isnull()
```

```python
df['diagnosis'].replace('B', 'Benign', inplace=True)
df['diagnosis'].replace('M', 'Malignant', inplace=True)
df
```

```python
feature_df = df.iloc[:,2:]
label_df =df.iloc[:,1]
```

```python
label_df
```

```python
feature_df
```

```python
from sklearn.preprocessing import StandardScaler

# x = feature_df.values
# x = StandardScaler().fit_transform(x)
x = StandardScaler().fit_transform(feature_df)
x
```

```python
import numpy as np

np.mean(x), np.std(x)
```

Apply PCA: how much variance is explained in this two components

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_df = pd.DataFrame(data=principalComponents_breast, columns=['PC1', 'PC2'])
```

```python
pca_breast.explained_variance_ratio_   #0.44272026 of the variance is explained in the first component
 #0.18971182 of the variance is explained in the second component
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

plt.figure()
plt.xlabel('PC1')
plt.ylabel('PC2')
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target,color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_breast_df.loc[indicesToKeep,'PC1'],
               principal_breast_df.loc[indicesToKeep,'PC2'],
               c = color)
```

```python
pca = PCA().fit(x)
evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}:{:0.4f}'.format(*k) for k in enumerate (evr)))
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
```

```python
data = pd.read_csv('./gisette_valid.data', header=None, delim_whitespace=True)
data_label = pd.read_csv("./gisette_valid.labels")
```

```python
from sklearn.preprocessing import StandardScaler

x_d = StandardScaler().fit_transform(feature_df_d)
x_d
```

```python
import numpy as np

np.mean(x_d), np.std(x_d)
```

Apply PCA

```python
from sklearn.decomposition import PCA

X_train_pca = PCA(n_components=2)
principalComponents = X_train_pca.fit_transform(x_d)
principalDf = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2'])
```

```python
X_train_pca.explained_variance_ratio_  
```

```python
pca = PCA().fit(x_d)
evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}:{:0.4f}'.format(*k) for k in enumerate (evr)))
```

```python

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

plt.figure()
plt.xlabel('PC1')
plt.ylabel('PC2')
targets = [1.0, -1.0]
colors = ['b', 'r']
for target,color in zip(targets, colors):
    indicesToKeep = label_df_d == target
    plt.scatter(principalDf.loc[indicesToKeep,'PC1'],
               principalDf.loc[indicesToKeep,'PC2'],
               c = color)
```
